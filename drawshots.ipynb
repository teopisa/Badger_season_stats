```python
import matplotlib.pyplot as plt
from mplbasketball import Court
import matplotlib
import json

# Imposta il backend per interattività
matplotlib.use('TkAgg')

# Configura il campo da basket
court = Court(court_type="nba", origin="center", units="m")
match = "season_stats/match_stats/colognola_badgers_20241031.json"

# Carica il file JSON contenente le statistiche dei giocatori
with open(match, "r") as f:
    data = json.load(f)

# Carica il file JSON contenente le statistiche dei giocatori
with open(match, "r") as f:
    data = json.load(f)

# Estrai i nomi dei giocatori dal campo 'statsgiocatori'
players = [player["name"] for player in data["stats_giocatore"]]
print(players)

# Dati tiri per ciascun giocatore
all_shots = {}

# Ciclo per ogni giocatore
for player in players:
    print(f"\nRegistrazione tiri per il giocatore: {player}")
    
    # Per ogni giocatore disegna il campo
    fig, ax = court.draw(showaxis=True, orientation="vu")
    plt.title(f"Tiri per {player} - Clicca sui punti di tiro")

    # Lista dei tiri del giocatore
    player_shots = []

    # Loop per cliccare punti senza limiti
    while True:
        try:
            print("Clicca sul campo per registrare un tiro oppure chiudi la finestra quando hai finito.")
            # Permette di selezionare un punto alla volta
            coord = plt.ginput(1, timeout=-1)[0]
            
            # Chiede se il tiro è stato segnato o sbagliato
            made = input("Il tiro è stato segnato? (s/n): ").strip().lower() == "s"

            # Salva il tiro con X, Y e risultato
            shot = {"X": coord[0], "Y": coord[1], "Made": made}
            player_shots.append(shot)

            # Visualizza il tiro sul campo: cerchio per segnati, X per sbagliati
            if made:
                ax.plot(coord[0], coord[1], 'o', color='green', markersize=8)  # Cerchio verde per i tiri segnati
            else:
                ax.plot(coord[0], coord[1], 'x', color='red', markersize=8)    # "X" rossa per i tiri sbagliati

            # Aggiorna il grafico
            plt.draw()

        except IndexError:
            # Uscita dal ciclo se chiudi la finestra
            break

    # Aggiunge i tiri del giocatore al dizionario principale
    all_shots[player] = player_shots
    plt.close(fig)  # Chiude la finestra dopo aver selezionato i punti

# Salva tutti i dati in un file JSON
with open("all_shots.json", "w") as f:
    json.dump(all_shots, f)

print("\nCoordinate dei tiri salvate in all_shots.json")
```
